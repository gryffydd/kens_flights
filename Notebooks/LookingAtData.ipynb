```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_file = "P753-line-magnetic-AWAGS_MAG_2010.dat"
df = pd.read_csv(data_file, delim_whitespace=True, header=None)
```

```python
def scatter(df, col=0):
    fig, ax = plt.subplots(figsize=(10,10))
    ax.scatter(df[10], df[9], c=df[col])
    plt.show()
```

```python
scatter(df[:50], col=3)
```
